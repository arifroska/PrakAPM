```python
import io
import numpy as np  #import library numpy
import pandas as pd  #import library pandas
import matplotlib.pyplot as plt  #import library maplotlib
%matplotlib inline
from sklearn import metrics  #import library metrics dari modul sklearn
from sklearn.metrics import accuracy_score, confusion_matrix  #import library metrics dari modul sklearn
from sklearn.model_selection import train_test_split  #import modul train test split dari library sklearn
import seaborn as sns  #import library seaborn

data = pd.read_csv(io.BytesIO(uploaded['survey lung cancer.csv']), encoding= 'latin-1')
data.head(n=10)
```

```python
# Import LabelEncoder
from sklearn import preprocessing

# Creating labelEncoder
le = preprocessing.LabelEncoder()

# Converting string labels into numbers.
data['LUNG_CANCER']=le.fit_transform(data['LUNG_CANCER'])
```

```python
# Spliting data into Feature and
X=data[['SMOKING', 'ANXIETY', 'FATIGUE']]
y=data['LUNG_CANCER']

# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)  # 70% training and 30% test
```

```python
# Import MLPClassifer 
from sklearn.neural_network import MLPClassifier

# Create model object
clf = MLPClassifier(hidden_layer_sizes=(6,5),
                    random_state=5,
                    verbose=True,
                    learning_rate_init=0.01)

# Fit data onto the model
clf.fit(X_train,y_train)
```

```python
# Make prediction on test dataset
ypred=clf.predict(X_test)

# Import accuracy score 
from sklearn.metrics import accuracy_score

# Calcuate accuracy
accuracy_score(y_test,ypred)
```
